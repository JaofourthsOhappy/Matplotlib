```python
import matplotlib.pyplot as plt

product1 = [10,20,30,40,50]
product2 = [15,30,10,20,60]
month = [1,2,3,4,5]
```

```python
plt.plot(month,product1,"ms--")
plt.plot(month,product2,"o:k")
plt.xlabel("Month",size=20,color="red",backgroundcolor="yellow")
plt.ylabel("Sales",size=21,color="blue",backgroundcolor="yellow")
plt.show
```
