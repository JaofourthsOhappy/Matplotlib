```python
import matplotlib.pyplot as plt

student = [10,40,30,25,60]
course = ["PHP","JAVA","PYTHON","C++","NODE JS"]
```

```python
colors = ["purple","red","cyan","yellow","green"]
exp = [0,0.5,0,0,0]
plt.pie(student,labels=course,autopct="%.1f%%",colors=colors,shadow=True,explode=exp)
plt.show()
```

```python
colors = ["purple","red","cyan","yellow","green"]
exp = [0,0.5,0,0,0]
plt.pie(student,labels=course,autopct="%.1f%%",colors=colors,shadow=True,explode=exp,startangle=90)
plt.show()
```
