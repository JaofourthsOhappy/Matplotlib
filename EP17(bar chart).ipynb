```python
import matplotlib.pyplot as plt

course = ["Python","PHP","C++"]
score = [80,75,50]
```

```python
colors = ["purple","red","cyan"]
plt.bar(course,score,color=colors,width=0.5,edgecolor="pink",linewidth=3)
plt.xlabel("Course")
plt.ylabel("Score")
plt.title("Score midterm")
plt.show()
```

```python
colors = ["purple","red","cyan"]
plt.barh(course,score,color=colors,edgecolor="pink",linewidth=3)
plt.xlabel("Score")
plt.ylabel("Course")
plt.title("Score midterm")
plt.show()
```
