```python
import matplotlib.pyplot as plt

# # มีห้องเรียน 3 ห้อง
# ่ในเเต่ละห้องมีนักเรียนชายหญิง
# เเสดงเป็นอัตราส่วนผ่าน stackbar
```

```python
room = ["A","B","C"]
boys = [10,15,5]
girls = [20,7,15]
```

```python
plt.bar(room,boys,label="Boys",color="blue")
plt.bar(room,girls,bottom=boys,label="Girls",color="red")
plt.show()
```

```python
plt.stackplot(room,boys,girls,labels=["Boys","Girls"],colors=["blue","pink"])
plt.legend()
```
