```python
import matplotlib.pyplot as plt

product1 = [10,20,30,40,50]
product2 = [15,30,10,20,60]
month = [1,2,3,4,5]
```

```python
plt.plot(month,product1,marker="x",linestyle="--")
plt.plot(month,product2,marker="o",linestyle=":")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show
```

```python
plt.plot(month,product1,"ms--")
plt.plot(month,product2,"o:k")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show
```
